```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
my_url = "https://tribecafilm.com/filmguide?letter=a"
raw_html = urlopen(my_url).read()
soup_doc = BeautifulSoup(raw_html, "html.parser")
```

```python
print(soup_doc.prettify())
```

```python
soup_doc.title
```

```python
all_movies = soup_doc.find_all('article')
```

```python
movie_info = all_movies[2].find('div', attrs={'class': 'left'})
movie_info.find_all('a')
```

```python
movie_info.find_all('li')
```

```python
for this_movie in all_movies:
    movie_info = this_movie.find('div', attrs={'class': 'left'})
    title = movie_info.find(class_="headerTitle").find('a').string
    print(title)
    details = movie_info.find_all('li')
    for dets in details[:-1]:
        print(dets.string)
    country = details[-1].find('a').string
    print(country)
```
